```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#import CSV file
data1 = pd.read_csv('ebolaData.csv', encoding='latin-1')
data1
```

```python
#select contact type 
```

```python
data2 = data1[data1.Gender=='Female']
```

```python
data2.head()
```

```python
data3 = data2[data2.DischType=="Died"]
```

```python
data3.head()
```

```python
sns.set(rc={'figure.figsize':(20,15)})
graph = sns.boxplot(x= "Fever", y="Age", data=data3, showfliers=False)
graph.axes.tick_params(axis='both', which='major', labelsize=30)
graph.set_xlabel("Fever", fontsize=40)
graph.set_ylabel("Age", fontsize=40)
graph.axes.set_title("Deaths Among Women Infected with Ebola Presenting with Fever as a Symptom",fontsize=40);
```

```python
#export graph
graph = graph.get_figure()
graph.savefig("FemaleDeaths_FeverSymptom.jpeg")
```
